```python
import numpy as np
import glob
import cv2
import os
from PIL import Image
import PIL.Image as im
import sewar.full_ref as swr
import imquality.brisque as brisque
from pathlib import Path
import skimage
from numpy import asarray
```

```python
datapath = Path.cwd()     #returns current working directory.  Still no idea why it's here
#datapath.joinpath("hello").joinpath("test") 
folder = r'/notebooks/TRIQ/Kadid/kadid10k/images'
files = glob.glob(os.path.join(folder,'*.png'))
files.sort()
```

```python
image_names = []
for file in files:
    name = os.path.splitext(os.path.basename(file))[0]
    name = name + '.png'
    image_names.append(name)
```

```python
def chunk_based_on_size(lst, n):
    for x in range(0, len(lst), n):
        each_chunk = lst[x: n+x]

        if len(each_chunk) < n:
            each_chunk = each_chunk + [None for y in range(n-len(each_chunk))]
        yield each_chunk
```

```python
imgGroups = list(chunk_based_on_size(image_names,126)) #2D array just like what we need. Zero indexed
```

```python
FolderPath = r'/notebooks/TRIQ/Kadid/kadid10k/images'
ImageFolder = Path(FolderPath)

#defining various lists
lst_images = []
lst_ref = []
lst_RMSE_SW = []
lst_MSE = []
lst_RMSE = []
lst_PSNR = []
lst_UQI = [] 
lst_SSIM = []
lst_ERGAS = []
lst_SCC = []
lst_RASE = []
lst_SAM = []
lst_VIFP = []
lst_MSSSIM = []
lst_PSNRB = []  
```

```python
for i in range(0,10):
    refImg = ImageFolder.joinpath(imgGroups[i][0])
    ref = im.open(refImg)
    nd1 = asarray(ref)
    #ref_width, ref_height = ref.size
    for j in range(0,126):
        try:
            lst_ref.append(imgGroups[i][0])
            lst_images.append(imgGroups[i][j])
            distortedImage = ImageFolder.joinpath(imgGroups[i][j])
            image = im.open(distortedImage)
            nd2 = asarray(image)
            #width, height = image.size
            lst_MSE.append(swr.mse(nd1,nd2))
            lst_RMSE.append(swr.rmse(nd1,nd2))
            lst_RMSE_SW.append(swr.rmse_sw(nd1,nd2))
            lst_PSNR.append(swr.psnr(nd1,nd2))
            lst_SSIM.append(swr.ssim(nd1,nd2))
            lst_ERGAS.append(swr.ergas(nd1,nd2))
            lst_SCC.append(swr.scc(nd1,nd2))
            lst_RASE.append(swr.rase(nd1,nd2))
            lst_SAM.append(swr.sam(nd1,nd2))
            lst_MSSSIM.append(swr.msssim(nd1,nd2))
            lst_VIFP.append(swr.vifp(nd1,nd2))
            lst_PSNRB.append(swr.psnrb(nd1,nd2))
            lst_UQI.append(swr.uqi(nd1,nd2))
            print("iteration: ",i,j)
        except:
            print("ERROR-----------iteration:",i,j)
```
